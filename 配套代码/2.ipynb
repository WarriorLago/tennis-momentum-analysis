```python
import pandas as pd

# 读取Excel文件
file_path_new = '第二问.xlsx'
full_data = pd.read_excel(file_path_new)

```

```python
full_data1=full_data
```

```python
# 重新调整计算逻辑以满足需求：当一方胜利时，另一方的连胜盘数归0

# 重置连胜盘数列
full_data1['本场比赛内球员1连胜盘数'] = 0
full_data1['本场比赛内球员2连胜盘数'] = 0

# 分组处理每场比赛
for match_id, group in full_data1.groupby('比赛ID'):
    # 初始化前一盘胜利情况
    prev_win_p1 = 0
    prev_win_p2 = 0

    # 初始化连胜盘数
    streak_p1 = 0
    streak_p2 = 0

    for i, row in group.iterrows():
        # 获取当前行的盘数胜利情况
        current_win_p1 = row['球员1赢得盘数']
        current_win_p2 = row['球员2赢得盘数']

        # 球员1胜利盘数变化，更新连胜盘数，球员2连胜盘数归0
        if current_win_p1 > prev_win_p1:
            streak_p1 += 1
            streak_p2 = 0  # 球员2连胜盘数归0
            prev_win_p1 = current_win_p1

        # 球员2胜利盘数变化，更新连胜盘数，球员1连胜盘数归0
        if current_win_p2 > prev_win_p2:
            streak_p2 += 1
            streak_p1 = 0  # 球员1连胜盘数归0
            prev_win_p2 = current_win_p2

        # 更新数据
        full_data1.at[i, '本场比赛内球员1连胜盘数'] = streak_p1
        full_data1.at[i, '本场比赛内球员2连胜盘数'] = streak_p2

```

```python
# 计算本场比赛内球员1连胜局数和球员2的连胜局数

# 初始化连胜局数列
full_data1['本场比赛内球员1连胜局数'] = 0
full_data1['本场比赛内球员2的连胜局数'] = 0

# 分组处理每场比赛
for match_id, group in full_data1.groupby('比赛ID'):
    # 初始化前一局球员1和球员2的胜局数
    prev_win_count_p1 = group.iloc[0]['球员1赢得局数'] - 1
    prev_win_count_p2 = group.iloc[0]['球员2赢得局数'] - 1

    # 初始化连胜局数
    streak_p1 = 0
    streak_p2 = 0

    for i, row in group.iterrows():
        # 获取当前行的局数胜利情况
        current_win_p1 = row['球员1赢得局数']
        current_win_p2 = row['球员2赢得局数']

        # 球员1局数胜利变化，更新连胜局数
        if current_win_p1 > prev_win_count_p1:
            streak_p1 += current_win_p1 - prev_win_count_p1
        else:
            streak_p1 = 0
        prev_win_count_p1 = current_win_p1

        # 球员2局数胜利变化，更新连胜局数
        if current_win_p2 > prev_win_count_p2:
            streak_p2 += current_win_p2 - prev_win_count_p2
        else:
            streak_p2 = 0
        prev_win_count_p2 = current_win_p2

        # 更新数据
        full_data1.at[i, '本场比赛内球员1连胜局数'] = streak_p1
        full_data1.at[i, '本场比赛内球员2的连胜局数'] = streak_p2

# 显示处理后的部分数据以验证计算逻辑
full_data1[['比赛ID', '球员1', '球员2', '球员1赢得局数', '球员2赢得局数', '本场比赛内球员1连胜局数', '本场比赛内球员2的连胜局数']].head(20)
```
