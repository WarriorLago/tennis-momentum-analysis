```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the data
data = pd.read_excel('Q2指标_带势头评分 -14点32分.xlsx')

# Set the style of the plots
sns.set(style="whitegrid")

# Create figure and axes for the subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Enhance the plot for Player 1
sns.regplot(data=data, x='选手1势头评分', y='球员1本局是否胜利', ax=axes[0], logistic=True, ci=95, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
axes[0].set_title('Player 1 Momentum Score vs. Match Victory (with Regression Line)')
axes[0].set_xlabel('Player 1 Momentum Score')
axes[0].set_ylabel('Player 1 Match Victory')
axes[0].set_ylim(-0.1, 1.1)

# Enhance the plot for Player 2
sns.regplot(data=data, x='选手2势头评分', y='球员2本局是否胜利', ax=axes[1], logistic=True, ci=95, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
axes[1].set_title('Player 2 Momentum Score vs. Match Victory (with Regression Line)')
axes[1].set_xlabel('Player 2 Momentum Score')
axes[1].set_ylabel('Player 2 Match Victory')
axes[1].set_ylim(-0.1, 1.1)

plt.tight_layout()
plt.show()
```
