```python
import pandas as pd

# Load the uploaded file
file_path = '测试3.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Initialize variables for tracking and storing metrics
segment_data = []
start_index = 0
prev_end_index = -1

# Iterate over the DataFrame to process each game segment
for i, row in data.iterrows():
    # Check if it's a new game segment or the last row
    if row['是否换局了'] == 1 or i == len(data) - 1:
        # Calculate metrics for the current segment
        if i == len(data) - 1 and row['是否换局了'] != 1:
            segment = data.iloc[start_index:]
        else:
            segment = data.iloc[start_index:i]
            start_index = i  # Update start index for the next segment
        
        # Calculate the sum and average of required metrics
        sum_unforced_errors = segment['失误次数（选手1-选手2）'].sum()
        sum_aces_diff = segment['制胜发球差'].sum()
        sum_winners_diff = segment['击球制胜球差'].sum()
        avg_performance_diff = segment['表现评分之差'].mean()
        
        # Append the calculated metrics to the list
        segment_data.append({
            '比赛ID': row['比赛ID'],
            '盘数': row['盘数'],
            '局数': row['局数'],
            '失误次数之和': sum_unforced_errors,
            '制胜发球差之和': sum_aces_diff,
            '击球制胜球差之和': sum_winners_diff,
            '表现评分之差平均值': avg_performance_diff
        })

        prev_end_index = i  # Update the end index for the previous segment

# Convert the list of dictionaries to a DataFrame
segment_summary = pd.DataFrame(segment_data)

# Display the first few rows of the summary DataFrame
segment_summary.head()
```

```python
output_file_path = '测试3算完的.xlsx'
segment_summary.to_excel(output_file_path, index=False)
```
